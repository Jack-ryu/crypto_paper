```python
import os 
import re

import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
path = 'Data/binance/binance_usdt'
data = pd.DataFrame()

for file_name in os.listdir(path):
    coin = file_name[:-4] 
    df = pd.read_csv(f'{path}/{file_name}')
    df['open_time'] = pd.to_datetime(df['open_time'], unit='ms')
    df['coin'] = coin
    data = pd.concat([data, df])
```

```python
data = data.sort_values(by='open_time').reset_index(drop=True)
data.loc[:, 'coin'] = data.coin.apply(lambda x: re.sub('_[0-9]+', '', x))
data
```

```python
pivot = pd.pivot_table(data,
                       values='volume',
                       index='open_time',
                       columns='coin')
coin_lst = pivot.columns
pp1 = [coin.replace('1000', '').replace('SETTLED','') for coin in coin_lst]
pp2 = [re.sub(r'(USDT|BUSD)$', '', item) for item in pp1]
pivot.columns = pp2
```

```python
pivot2 = pivot.loc[:, ~pivot.columns.duplicated()]
pivot2
```
